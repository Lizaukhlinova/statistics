Построим график зависимости модуля разности оценки дисперсии и ее истинного значения от $n$.

```python
def difference (func, sample):
    y = np.array([], dtype=float)
    for k in range(1, size_N):
        y = np.append(y, abs(func(sample[:k]) - 5. / 48.))
    return y

n = np.arange(1, size_N, dtype=int)
plt.plot(n, difference(S, sample), 'r')
plt.ylim([0, 0.10])
plt.show()
```

Проведем аналогичное исследование для выборки из распределения Коши. Построим график зависимости оценки дисперсии от $n$.

```python
sample_cauchy = st.cauchy().rvs(size=size_N)
y = np.array([], dtype=float)
for k in range(1, size_N):
    y = np.append(y, S(sample_cauchy[:k]))
plt.plot(n, y, 'g')
plt.show()
```
